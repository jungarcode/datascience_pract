## EDA Heart-Disease(Enfermo-corazon)

Este conjunto de datos contiene los registros médicos de 302 pacientes que tuvieron insuficiencia cardíaca, recopilados durante un período de seguimiento, donde cada perfil de paciente tiene 13 características clínicas.

Se establecera un modelo que permita calcular la probabilidad de que el paciente muera durante un periodo de tiempo, La variable Target está codificada como 1 si el paciente esta enfermo y 0 si no lo esta.

**Informacion Atributos:**

[https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset](https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset)

> * Age (age in years)
> * Sex (1 = male; 0 = female)
> * CP (chest pain type)
> * TRESTBPS (resting blood pressure (in mm Hg on admission to the hospital))
> * CHOL (serum cholestoral in mg/dl)
> * FPS (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
> * RESTECH (resting electrocardiographic results)
> * THALACH (maximum heart rate achieved)
> * EXANG (exercise induced angina (1 = yes; 0 = no))
> * OLDPEAK (ST depression induced by exercise relative to rest)
> * SLOPE (the slope of the peak exercise ST segment)
> * CA (number of major vessels (0-3) colored by flourosopy)
> * THAL (3 = normal; 6 = fixed defect; 7 = reversable defect)
> * TARGET (1 or 0)



### Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
import warnings
warnings.filterwarnings('ignore')
```

### Cargamos Dataset

```python
falla_cardiaca = pd.read_csv('heart.csv')
falla_cardiaca.head()
```

### Informacion de Dataset

```python
falla_cardiaca.info()
```

### Valores Nulos

```python
falla_cardiaca.isna().sum().sort_values(ascending=False)
```

### Valores Duplicados

```python
falla_cardiaca.duplicated().sum()
```

### Eliminamos Valores Duplicados

```python
falla_cardiaca = falla_cardiaca.drop_duplicates()
```

```python
falla_cardiaca.shape
```

### Estadisticas del Dataset

```python
falla_cardiaca.describe().T
```

### Cantidad de Pacientes con enfermedad en el corazon

```python
falla_cardiaca.target.value_counts().sort_index()
```

```python
sns.countplot(
    falla_cardiaca['target']
)
plt.xticks([0,1],['No-Enfermo','Enfermo'])
plt.title('Cantidad de Pacientes con Enfermedad Corazon')
plt.show()
```

### Matriz de Correlacion

```python
falla_cardiaca.corr()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(
    falla_cardiaca.corr(),
    cmap='Blues',
    #square=True,
    linewidth=0.1,
    annot= True
)
plt.show()
```

### Cantidad de Pacientes Enfermos por Genero

```python
sns.countplot(
    data=falla_cardiaca,
    x='sex',
    hue='target',  
)
plt.xticks([0,1],['Mujeres','Hombres'])
plt.legend(labels=['No-Enfermo','Enfermo'])
plt.title('Cantidad de Pacientes x Genero')
plt.show()
```

### Distribucion del Dataset x Edad

```python
sns.distplot(
    falla_cardiaca['age'],
    bins=20,
    kde=True
).set(title='Distribucion x Edad')
plt.show()
```

### Como influye la variable Chest pain(dolor de pecho) en los Pacientes

```python
sns.countplot(
    data=falla_cardiaca,
    x='cp',
    hue='target'
)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal pain','asymtomatic'])
plt.legend(labels=['No-Enfermo','Enfermo'])
plt.title('Influencia CP en Pacientes Enfermos')
plt.show()
```

### Como influye la variable FBS(glucemia en ayunas) en los Pacientes

```python
sns.countplot(
    data=falla_cardiaca,
    x='fbs',
    hue='target'
)
plt.xticks([0,1],['False','True'])
plt.legend(labels=['No-Enfermo','Enfermo'])
plt.title('Influencia FBS en Pacientes Enfermos')
plt.show()
```
