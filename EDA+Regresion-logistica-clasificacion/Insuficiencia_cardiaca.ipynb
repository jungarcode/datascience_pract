## EDA Heart Disease

Este conjunto de datos contiene los registros médicos de 302 pacientes que tuvieron insuficiencia cardíaca, recopilados durante un período de seguimiento, donde cada perfil de paciente tiene 13 características clínicas.

Se establecera un modelo que permita calcular la probabilidad de que el paciente muera durante un periodo de tiempo, La variable Target está codificada como 1 si el paciente muere y 0 si sobrevive.

**Informacion Atributos:**

[https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset](https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset)

> **Attribute Information**
> * Age (age in years)
> * Sex (1 = male; 0 = female)
> * CP (chest pain type)
> * TRESTBPS (resting blood pressure (in mm Hg on admission to the hospital))
> * CHOL (serum cholestoral in mg/dl)
> * FPS (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
> * RESTECH (resting electrocardiographic results)
> * THALACH (maximum heart rate achieved)
> * EXANG (exercise induced angina (1 = yes; 0 = no))
> * OLDPEAK (ST depression induced by exercise relative to rest)
> * SLOPE (the slope of the peak exercise ST segment)
> * CA (number of major vessels (0-3) colored by flourosopy)
> * THAL (3 = normal; 6 = fixed defect; 7 = reversable defect)
> * TARGET (1 or 0)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
import warnings
warnings.filterwarnings('ignore')
```

### Cantidad de Pacientes con enfermedad en el corazon

```python
falla_cardiaca.target.value_counts().sort_index()
```

```python
sns.countplot(
    falla_cardiaca['target']
)
plt.xticks([0,1],['No-Enfermo','Enfermo'])
plt.title('Cantidad de Pacientes con Enfermedad Corazon')
plt.show()
```

### Matriz de Correlacion

```python
falla_cardiaca.corr()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(
    falla_cardiaca.corr(),
    annot= True
)
```

### Cantidad de Pacientes Enfermos por Genero

```python
sns.countplot(
    data=falla_cardiaca,
    x='sex',
    hue='target',  
)
plt.xticks([0,1],['Mujeres','Hombres'])
plt.legend(labels=['No-Enfermo','Enfermo'])
plt.title('Cantidad de Pacientes x Genero')
plt.show()
```
